```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
%matplotlib inline
```

```python
df_comp = pd.read_csv('./Final Scoresheet NZB - Sheet1.csv')  
df_comp.head(5)
```

```python
data = []
```

```python
student_count = 0;

for i, student_id in enumerate(df_comp['Student ID']):
    if not pd.isna(student_id):
        data_per_student = {
            "student_id": df_comp['Student ID'][i],
            "Email": df_comp['Email'][i],
            "Internet Research": df_comp['Internet Research'][i],
            "Google Tools": df_comp['Google Tools'][i],
            "Documentation": df_comp['Documentation'][i],
            "Data Analysis": df_comp['Data Analysis'][i],
            "Presentation": df_comp['Presentation'][i]
        }
        data.append(data_per_student)
# print(data)
```

```python


# print(data.head(3))
```

```python
#Plotting Trends

for data_ in data:
    print(data_)
#     student_id = "x"
    student_id = data_["student_id"]
    del data_["student_id"]
#     print(data_)
    count_df = pd.Series(data_).to_frame('Score')


    # Resetting Index 
    count_df.index.name = 'Name'

    count_df.reset_index(drop=False, inplace=True)
    print(count_df)

    plotTitle = ''

    sns.set(style="white", rc={"lines.linewidth": 3})

    fig, ax = plt.subplots(figsize=(12, 5))

    ax.set_title(plotTitle, fontdict={'fontsize':12,'fontweight':'bold'})
    # ax.set_axisbelow(True)
    ax.xaxis.grid(color='gray', linestyle='dashed')
    # ax.hlines()
    sns.barplot(x='Score',
                y='Name',
                data=count_df,
                color='#3592f0',
                ax=ax,
                )

    ax.set_ylabel('')

    sns.despine()

    sns.set()


    # # Saving plot to output .png file

    plt.savefig('./outputs/nzb/{}.png'.format(student_id), dpi=100)
    del count_df
    del fig
    del ax
```
