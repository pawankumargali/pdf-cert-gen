```python
import fitz  # PyMuPDF

# Open the PDF file
pdf_path = "template.pdf"  # Replace with your PDF path
doc = fitz.open(pdf_path)

# Loop through each page
for page_num in range(len(doc)):
    page = doc[page_num]
    print(f"\nPage {page_num + 1}:")

    # Extract text and character positioning with font info
    for text_block in page.get_text("dict")["blocks"]:
        if "lines" in text_block:  # Text blocks contain 'lines'
            for line in text_block["lines"]:
                for span in line["spans"]:
                    text = span["text"]            # Text content
                    font = span["font"]            # Font name
                    size = span["size"]            # Font size
                    color = span["color"]          # Font color (as an integer)
                    bbox = span["bbox"]            # Bounding box of the text (x0, y0, x1, y1)

                    # Print the extracted information
                    print(f"Text: '{text}'")
                    print(f"Font: {font}, Size: {size}")
                    print(f"Color (RGB): {color}")
                    print(f"Position: {bbox}")
                    print("-" * 50)

doc.close()
```

```python
import fitz
# Load the PDF file
pdf_path = "./template.pdf"  # Replace with the path to your PDF
doc = fitz.open(pdf_path)

# Extract text and positions from the first page
page = doc[0]
text_instances = []

# Search for "<Student Name>" placeholder
search_text = "<Student Name>"
for inst in page.search_for(search_text):
    x0, y0, x1, y1 = inst
    center_x, center_y = (x0 + x1) / 2, (y0 + y1) / 2
    text_instances.append({
        "text": search_text,
        "position": (x0, y0, x1, y1),
        "center": (center_x, center_y)  # Center of the bounding box
    })

doc.close()
print(text_instances)


```
